## 1.Most viewed videos on Youtube

```python
#initialising the webdriver
driver=webdriver.Chrome(r"I:\\chromedriver_win32\\chromedriver.exe")

#accesing the website
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
time.sleep(2)

rank=[]
name=[]
artist=[]
upload=[]
views=[]


#Scrolldown the page
#driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")

#Getting the rank
num=driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[1]')
for i in num:
    rank.append(i.text)

#Getting the name of video
video_names=driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[2]')
for video in video_names:
    name.append(video.text)
    
#Getting the artist details
uploaders=driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[3]')
for uploader in uploaders:
    artist.append(uploader.text)
    
#Getting the upload date
uploaded_on=driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[5]')
for uploaded in uploaded_on:
    upload.append(uploaded.text)

#Getting the views
tot_views=driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[4]')
for view in tot_views:
    views.append(view.text)
    

driver.close()

youtube=pd.DataFrame({'Rank':rank,'Name':name,'Artist':artist,'Uploaded On':upload, 'Views in billions': views})
youtube


```

## 2. Details from BCCI

```python
#initialising the webdriver
driver=webdriver.Chrome(r"I:\\chromedriver_win32\\chromedriver.exe")

#accesing the website
driver.get('https://www.bcci.tv/')


#accessing the international fixtures
i_fixtures=driver.find_element_by_xpath('//div[@class="navigation__drop-down drop-down drop-down--reveal-on-hover"]/div/ul/li[1]/a')
link=i_fixtures.get_attribute('href')
driver.get(link)

title=[]
series=[]
place=[]
date=[]
time=[]
link=[]

#button=driver.find_element_by_xpath('/html/body/div[4]/div/div/div[2]/section/button')
#button.click()


#Getting the links
links=driver.find_elements_by_xpath('//div[@class="event-list__list"]/div[1]/a')
try:
    for j in links:
        link.append(j.get_attribute('href'))
except:
    link.append('-')

for i in link:
    driver.get(i)
    driver.implicitly_wait(2)
    
    #Getting the title of match
    try:
        titles=driver.find_element_by_xpath('//span[@class="mc-header-info__match-description-match-type mc-theme-color"]')
        title.append(titles.text)
    except:
        title.append('-')
        
    #Getting date
    try:
        date_data=driver.find_element_by_xpath('//div[@class="mc-header-scorebox__datetime"]/strong[1]')
        date.append(date_data.text)
    except:
        date.append('-')
    
    #Getting time
    try:
        time_data=driver.find_element_by_xpath('//span[@class="mc-header-scorebox__ist-time"]')
        time.append(time_data.text)
    except:
        time.append('-')
        
    #Getting place
    try:
        places=driver.find_element_by_xpath('//*[@id="scorecard"]/section[3]/div/ul/li[3]/span[2]')
        place.append(places.text)
    except:
        place.append('-')
        
        
bcci=pd.DataFrame({'Title':title,'Place':place,'Date':date,'Time':time})
bcci
```

##  4. State-wise GDP INDIA

```python
#initialising the webdriver
driver=webdriver.Chrome(r"I:\\chromedriver_win32\\chromedriver.exe")

#accesing the website
driver.get('https://www.statisticstimes.com/')

#Getting the link of economy
economy=driver.find_element_by_xpath('//div[@class="navbar"]/div[2]/div/a[3]')
link=economy.get_attribute('href')

driver.get(link)

#States gdp link
states_gdp=driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[2]/ul/li[1]/a')
stateslink=states_gdp.get_attribute('href')

driver.get(stateslink)

rank=[]
state=[]
gsdp19_20=[]
gsdp18_19=[]
share=[]
gdp=[]

ranks=driver.find_elements_by_xpath('//*[@id="table_id"]/tbody/tr/td[1]')
for i in ranks:
    rank.append(i.text)
    

states=driver.find_elements_by_xpath('//*[@id="table_id"]/tbody/tr/td[2]')
for j in states:
    state.append(j.text)
    
gsdp_19_20=driver.find_elements_by_xpath('//*[@id="table_id"]/tbody/tr/td[3]')
for k in gsdp_19_20:
    gsdp19_20.append(k.text)
    
gsdp_18_19=driver.find_elements_by_xpath('//*[@id="table_id"]/tbody/tr/td[4]')
for b in gsdp_18_19:
    gsdp18_19.append(b.text) 
    
shares=driver.find_elements_by_xpath('//*[@id="table_id"]/tbody/tr/td[5]')
for s in shares:
    share.append(s.text) 
    
gdps=driver.find_elements_by_xpath('//*[@id="table_id"]/tbody/tr/td[6]')
for g in gdps:
    gdp.append(g.text) 

state_wise_gdp=pd.DataFrame({'Rank':rank, 'State':state, 'GDP(19-20)':gsdp19_20, 'GDP(18_19)':gsdp18_19, 'Share':share, 'GDP ($ billions)':gdp})
state_wise_gdp
```

## 5. Details from github

```python
#initialising the webdriver
driver=webdriver.Chrome(r"I:\\chromedriver_win32\\chromedriver.exe")

#accesing the website
driver.get('https://github.com/')

#Getting link of trending
trending_link=driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a')
link=trending_link.get_attribute('href')
driver.get(link)

repos=driver.find_elements_by_xpath('//div[@class="position-relative container-lg p-responsive pt-6"]/div[1]/div[2]/article/h1')
repo_name=[repo.text for repo in repos]

repos_link=driver.find_elements_by_xpath('//div[@class="position-relative container-lg p-responsive pt-6"]/div[1]/div[2]/article/h1/a')
repo_link=[repos.get_attribute('href') for repos in repos_link]

des=[]
count=[]
language=[]

for links in repo_link:
    driver.get(links)
    
    #Getting description
    try:
        description=driver.find_element_by_xpath('//div[@class="BorderGrid-row hide-sm hide-md"]/div[1]/p')
        des.append(description.text)
    except NoSuchElementException:
        des.append('-')
        
    #Getting contributor count
    try:
        contributors=driver.find_element_by_xpath('//div[@class="BorderGrid BorderGrid--spacious"]/div[4]/div/h2/a/span')
        count.append(contributors.text)
    except NoSuchElementException:
        count.append('-')
        
    try:
        languages=driver.find_element_by_xpath('//div[@class="BorderGrid BorderGrid--spacious"]/div[5]/div[1]/ul/li[1]')
        language.append(languages.text.replace('\n',' '))
    except NoSuchElementException:
        language.append('-')
        
driver.close()

github=pd.DataFrame({'Title':repo_name, 'Description':des, 'Contributorscount':count, 'Language':language})
github
```

## 6. Details from Bilboard

```python
#initialising the webdriver
driver=webdriver.Chrome(r"I:\\chromedriver_win32\\chromedriver.exe")

#accesing the website
driver.get('https://www.billboard.com/')

#Clicking on full list
full_chart=driver.find_element_by_xpath('//div[@class="chart-promo__button-wrapper text--right"]/a')
link=full_chart.get_attribute('href')
driver.get(link)

song=[]
artist=[]
last_week=[]
peak=[]
weeks=[]

#Getting the song name
song_name=driver.find_elements_by_xpath('//span[@class="chart-element__information__song text--truncate color--primary"]')
for songs in song_name:
    song.append(songs.text)
    
#Getting the artist
artist_name=driver.find_elements_by_xpath('//span[@class="chart-element__information__artist text--truncate color--secondary"]')
for artists in artist_name:
    artist.append(artists.text)
    
#last week rank
last_week_rank=driver.find_elements_by_xpath('//div[@class="chart-element__meta text--center color--secondary text--last"]')
for last in last_week_rank:
    last_week.append(last.text)
    
#peak rank
peak_rank=driver.find_elements_by_xpath('//div[@class="chart-element__meta text--center color--secondary text--peak"]')
for ranks in peak_rank:
    peak.append(ranks.text)
    
#weeks on board
weeks_on_board=driver.find_elements_by_xpath('//div[@class="chart-element__meta text--center color--secondary text--week"]')
for weeks_on in weeks_on_board:
    weeks.append(weeks_on.text)
    
billboard=pd.DataFrame({'Song Name':song, 'Artist':artist, 'Last week Rank':last_week, 'Peak Rank':peak, 'Weeks On Board':weeks})
billboard
```

## 7. Data Science Recruiters from Naukri

```python
#initialising the webdriver
driver=webdriver.Chrome(r"I:\\chromedriver_win32\\chromedriver.exe")

#accesing the website
driver.get('https://www.naukri.com/')

#Getting recruiters
recruiters_link=driver.find_element_by_xpath('//a[@title="Search Recruiters"]')
recruiters_link.click()

#Switching tab
driver.switch_to.window(driver.window_handles[1])

#Finding the search and send keys
search_bar=driver.find_element_by_class_name('sugInp')
search_bar.send_keys('Data Science')

#Clicking the find button
buttons=driver.find_element_by_id('qsbFormBtn')
buttons.click()


name=[]
designation=[]
company=[]
skill=[]
location=[]

#Name of recruiter
try:
    recruiters=driver.find_elements_by_xpath('//p[@class="highlightable"]/a[1]/span[1]')
    for names in recruiters:
        name.append(names.text)
except NoSuchElementException:
    name.append('-')

#Designation of recruiter
try:
    designation_recruiter=driver.find_elements_by_xpath('//span[@class="ellipsis clr"]')
    for designations in designation_recruiter:
        designation.append(designations.text)
        
except NoSuchElementException:
    designation.append('-')

#Company
try:
    company_name=driver.find_elements_by_xpath('//p[@class="highlightable"]/a[2]')
    for companies in company_name:
        company.append(companies.text)
except NoSuchElementException:
    company.append('-')
    
#skills hiring
try:
    skills_hiring=driver.find_elements_by_xpath('//div[@class="hireSec highlightable"]')
    for skills in skills_hiring:
        skill.append(skills.text)
except NoSuchElementException:
    skill.append('-')

#location
try:
    location_details=driver.find_elements_by_xpath('//p[@class="highlightable"]/span[2]')
    for locations in location_details:
        location.append(locations.text)
except NoSuchElementException:
    location.append('-')
location.append('-')
    
Recruiters=pd.DataFrame({'Recruiter':name, 'Designation':designation, 'Company':company, 'Skills Hiring For':skill, 'Location':location})
Recruiters

```

## 10. UCI datasets

```python
#initialising the webdriver
driver=webdriver.Chrome(r"I:\\chromedriver_win32\\chromedriver.exe")

#accesing the website
driver.get('https://archive.ics.uci.edu/ml/index.php')

#Clicking on view all
view_all=driver.find_element_by_xpath('//span[@class="normal"]/b/a')
link=view_all.get_attribute('href')
driver.get(link)

#Data
datasets=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr')
data=[dt.text.split('\n') for dt in datasets]
dataset=[data[i][0] for i in range(len(data))]

#Data type
datatypes=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]')
datatype=[da.text for da in datatypes]

#Task
default_task=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]')
task=[t.text for t in default_task]

#attribute type
attributes=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]')
attr=[at.text for at in attributes]

#Instances
instances_num=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]')
instance=[ins.text for ins in instances_num]

#no. attribute
num_attribute=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]')
num=[n.text for n in num_attribute]

#year
years=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]')
year=[y.text for y in years]

driver.close()

UCI=pd.DataFrame({'Dataset':dataset[1:],'Datatypes':datatype[1:],'Task':task[1:],'Attributes':attr[1:],'Instances':instance[1:],
                  'No. of Attributes':num[1:], 'Year':year[1:]})
UCI
```
